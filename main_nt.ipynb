```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados
df = pd.read_csv('dados_jardim_das_flores.csv')

# Visualizar as primeiras linhas do dataframe
df.head()
```

```python
# Limpeza dos dados
# Verificar valores nulos
print(df.isnull().sum())

# Remover valores nulos
df.dropna(inplace=True)
```

```python
# Análise básica
# Distribuição de renda
plt.figure(figsize=(10,6))
sns.histplot(df['renda'], bins=20, kde=True)
plt.title('Distribuição de Renda')
plt.xlabel('Renda')
plt.ylabel('Frequência')
plt.savefig('distribuicao_renda.png')
plt.show()
```

```python
# Nível de escolaridade
escolaridade_counts = df['escolaridade'].value_counts()
plt.figure(figsize=(10,6))
escolaridade_counts.plot(kind='bar')
plt.title('Nível de Escolaridade')
plt.xlabel('Escolaridade')
plt.ylabel('Número de Pessoas')
plt.savefig('nivel_escolaridade.png')
plt.show()
```

```python
# Faixa etária
faixa_etaria_counts = df['faixa_etaria'].value_counts().sort_index()
plt.figure(figsize=(10,6))
faixa_etaria_counts.plot(kind='bar')
plt.title('Distribuição por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Pessoas')
plt.savefig('distribuicao_faixa_etaria.png')
plt.show()
```
